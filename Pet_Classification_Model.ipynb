**The CNN model should have the following layers:**  
● Input layer  
● Convolutional layer 1 with 32 filters of kernel size[5,5]  
● Pooling layer 1 with pool size[2,2] and stride 2  
● Convolutional layer 2 with 64 filters of kernel size[5,5]  
● Pooling layer 2 with pool size[2,2] and stride 2  
● Dense layer whose output size is fixed in the hyper parameter: fc_size=32 

● Dropout layer with dropout probability 0.4  
Predict the class by doing a softmax on the output of the dropout layers.  
This should be followed by training and evaluation:  
● For the training step, define the loss function and minimize it  
● For the evaluation step, calculate the accuracy

```python
train_path ='/content/drive/MyDrive/AI Engineering Labs/Deep Learning with Keras and Tensorflow/Final Projects/pet classification data/pet classification data/train'
test_path = '/content/drive/MyDrive/AI Engineering Labs/Deep Learning with Keras and Tensorflow/Final Projects/pet classification data/pet classification data/test'
```

```python
from keras import backend as k
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Dense, BatchNormalization, Flatten
```

```python
k.set_image_data_format('channels_last')
```

```python
model = Sequential()
```

```python
# 32 filters of 3*3 filter size
model.add(Conv2D(32,(5,5), input_shape = (150,150,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2))) #First Layer Completed

model.add(Conv2D(64,(5,5)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2))) #Second Layer Completed

model.add(Flatten())
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='softmax'))

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = 'accuracy')
```

```python
model.summary()
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator( rescale= 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```python
test_datagen = ImageDataGenerator(rescale= 1./255)
```

```python
train_generator = train_datagen.flow_from_directory(train_path,
                                                    target_size = (150, 150),
                                                    batch_size = 16,
                                                    class_mode= 'binary')
```

```python
test_generator = test_datagen.flow_from_directory(test_path,
                                                    target_size = (150, 150),
                                                    batch_size = 16,
                                                    class_mode= 'binary')
```

```python
model.fit_generator(train_generator,
                    epochs = 100,
                    validation_data = test_generator)
```

```python
model.fit_generator(train_generator,
                    epochs = 200,
                    validation_data = test_generator)
```

```python
model.fit_generator(train_generator,
                    epochs = 300,
                    validation_data = test_generator)
```
